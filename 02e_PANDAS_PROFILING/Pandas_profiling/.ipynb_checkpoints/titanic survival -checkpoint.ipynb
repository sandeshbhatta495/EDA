# Feature Engineering 

```python
# working With family size (adding the extra colum on family size )
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
```

```python
#check
df.head
```

```python
# replace family size 1 for categroical values 'alone ' like he/she travelling alone check if alone value is 1 
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)
```

# handling missing values 

```python
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
```

```python
# replacing the null values of age by its median 
df["Age"] = df.groupby("Title")["Age"].transform(lambda x: x.fillna(x.median()))
```

```python
df.head(5)
```

```python
# Fill Embarked with mode
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
#Fill Fare based on Pclass median
df["Fare"] = df.groupby("Pclass")["Fare"].transform(lambda x: x.fillna(x.median()))
```

```python
#Cabin  Extract Deck
df["Cabin"] = df["Cabin"].fillna("Unknown")
df["Deck"] = df["Cabin"].str[0]
df = df.drop("Cabin", axis=1)
```

```python
# creating Age category
df["AgeGroup"] = pd.cut(df["Age"],
                        bins=[0, 12, 18, 35, 60, 100],
                        labels=['Child', 'Teen', 'Young', 'Adult', 'Senior'])
# repacing the nummerical value with categorical values  leaving mininimal options 
```

```python
# creating the fare category 
df["FareGroup"] = pd.qcut(df["Fare"], 4)
```

```python
# dropping the useless column which has thier unique identity (not use for model traning )
df = df.drop(["PassengerId", "Name", "Ticket"], axis=1)
```

```python
# cncode the remnaing caterocial column
df = pd.get_dummies(df, columns=[
    "Sex", "Embarked", "Title", "Deck", "AgeGroup", "FareGroup"
], drop_first=True)
# like creating the many options and checking the each option and if true gets else 0 for false.
```

```python
# identify and represents the nummerical which need to use on the model
numeric_cols = ["Age", "Fare", "FamilySize"]
```
